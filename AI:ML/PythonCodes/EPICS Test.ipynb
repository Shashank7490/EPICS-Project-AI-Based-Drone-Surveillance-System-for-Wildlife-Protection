```python
import os

# ======== CONFIG ========
dataset_path = "/Users/shashankk/Desktop/Data Science and AIML/EPICS/Tortoise Detection Yolov8 Augmented/"   # Path to your YOLOv8 dataset
start_class_id = 0                  # The new class ID you want for this dataset

# Assuming YOLOv8 structure: dataset/labels/train/ (and optionally /val, /test)
label_folders = [
    os.path.join(dataset_path, "train/labels"),
    os.path.join(dataset_path, "val/labels"),
    os.path.join(dataset_path, "test/labels")
]

for label_dir in label_folders:
    if not os.path.exists(label_dir):
        continue

    for label_file in os.listdir(label_dir):
        if not label_file.endswith(".txt"):
            continue

        label_path = os.path.join(label_dir, label_file)

        with open(label_path, "r") as f:
            lines = f.readlines()

        new_lines = []
        for line in lines:
            parts = line.strip().split()
            if len(parts) > 0:
                parts[0] = str(start_class_id)  # change only the class ID
                new_lines.append(" ".join(parts) + "\n")

        with open(label_path, "w") as f:
            f.writelines(new_lines)

print(f"âœ… All labels in {dataset_path} updated to class ID {start_class_id}")
```
